# feature Encoding 
- If feature are present in binary form then use **LAbel encoding** here [time, smoker, sex are binary outcomes so use LAble encoding]
- If feature are present more than two then use **OneHot Encoding** [ day feature consit of more than two outcome so use OneHot Encoding)

```python
# Independent and dependent features 
df.columns
```

```python
X = df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y = df['total_bill']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.25 ,random_state=0)
```

```python
X_train.head()
```

# LAbel encoding 

```python
# LAbel encoding 
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex=LabelEncoder()
le_sex.fit(X_train['sex'])

le_smoker=LabelEncoder()
le_smoker.fit(X_train['smoker'])

le_time=LabelEncoder()
le_time.fit(X_train['time'])
```

```python
# Transform the training data
X_train['sex'] = le_sex.transform(X_train['sex'])
X_train['smoker'] = le_smoker.transform(X_train['smoker'])
X_train['time'] = le_time.transform(X_train['time'])
```

```python
X_train.head()
```

```python
# Transform the test data
X_test['sex'] = le_sex.transform(X_test['sex'])
X_test['smoker'] = le_smoker.transform(X_test['smoker'])
X_test['time'] = le_time.transform(X_test['time'])
```

```python
X_test.head()
```

```python
# OneHot Encoding for More than two outcome -- columnTransformer 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct=ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[3])],remainder='passthrough')
```

```python
ct.fit_transform(X_train)
```

```python
import sys
np.set_printoptions(threshold=sys.maxsize)
X_train = ct.fit_transform(X_train)
```

```python
X_test=ct.fit_transform(X_test)
```

# Support Vector Regressor 

```python
from sklearn.svm import SVR

# Train the SVR model on the preprocessed training data
svr = SVR(kernel='rbf')
svr.fit(X_train, y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.svm import SVR

# Train the SVR model on the preprocessed training data
svr = SVR(kernel='poly')
svr.fit(X_train, y_train)
```

```python
y_pred1 = svr.predict(X_test)
```

```python
print(r2_score(y_test,y_pred1))
print(mean_absolute_error(y_test,y_pred1))
```

```python
from sklearn.svm import SVR

# Train the SVR model on the preprocessed training data
svr = SVR(kernel='linear')
svr.fit(X_train, y_train)
```

```python
y_pred2 = svr.predict(X_test)
```

```python
print(r2_score(y_test,y_pred2))
print(mean_absolute_error(y_test,y_pred2))
```

# Hyperparameter tuning with GridSeachCV 

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, r2_score

# Define parameter range
param_grid = {
    'C': [0.01, 0.1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf']
}
```

```python
# Create GridSearchCV object
grid = GridSearchCV(SVR(), param_grid=param_grid, cv=5, refit=True, verbose=3)
```

```python
# Fit the grid search to the data
grid.fit(X_train, y_train)
```

```python
# Get the best parameters
best_params = grid.best_params_
print("Best parameters found: ", best_params)
```

```python
# Prediction
y_pred4 = grid.predict(X_test)
```

```python
# Regression metrics
mse = mean_absolute_error(y_test, y_pred4)
r2 = r2_score(y_test, y_pred4)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

# Hyperparameter tuning RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVR
from scipy.stats import uniform

param_dist = {
    'C': uniform(0.01, 100),
    'gamma': uniform(0.001, 1),
    'kernel': ['rbf']
}
```

```python
random_search = RandomizedSearchCV(SVR(), param_distributions=param_dist, n_iter=100, cv=5, verbose=3)
```

```python
random_search.fit(X_train, y_train)
```

```python
print("Best parameters:", random_search.best_params_)
```

```python
# Prediction
y_pred5 = random_search.predict(X_test)
```

```python
# Regression metrics
mae = mean_absolute_error(y_test, y_pred5)
r2 = r2_score(y_test, y_pred5)

print(f"Mean Squared Error: {mae}")
print(f"R^2 Score: {r2}")
```

# Hyperparameter with Optuna

```python
!pip install optuna
import optuna
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
```

```python
def objective(trial):
    C = trial.suggest_loguniform('C', 0.01, 100)
    gamma = trial.suggest_loguniform('gamma', 0.001, 1)
    kernel = trial.suggest_categorical('kernel', ['rbf'])
    
    svr = SVR(C=C, gamma=gamma, kernel=kernel)
    score = cross_val_score(svr, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    return score.mean()
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)
```

```python
print("Best parameters:", study.best_params)
```

```python
# Train the SVR model with the best parameters
best_svr = SVR(**study.best_params)
best_svr.fit(X_train, y_train)
```

```python
# Predict on the test data
y_pred7 = best_svr.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred7)
r2 = r2_score(y_test, y_pred7)

print(f"Mean Squared Error: {mse}")
print(f"RÂ² Score: {r2}")
```
